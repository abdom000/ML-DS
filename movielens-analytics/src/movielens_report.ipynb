# MovieLens Analytics Report

This report analyzes a 1000-record sample of the MovieLens dataset using only the classes and methods implemented in `movielens_analysis.py`.  
Each code cell includes the `%timeit` magic to measure the execution time of the methods.

```python
from movielens_analysis import Movies, Ratings, Tags, Links, MOVIES_1000, RATINGS_1000, TAGS_1000, LINKS_1000

# Create main objects
movies = Movies(MOVIES_1000)
ratings = Ratings(RATINGS_1000)
tags = Tags(TAGS_1000)
links = Links(LINKS_1000)

# Measure initialization time of the Movies class
%timeit Movies(MOVIES_1000)
```

## Movies: release years and genres

```python
# Distribution by release year
release_dist = movies.dist_by_release()
print("Distribution of movies by release year (sample):")
for year, count in list(release_dist.items())[:10]:
    print(year, ":", count)

# Distribution by genres
genre_dist = movies.dist_by_genres()
print("\nTop 10 genres:")
for genre, count in list(genre_dist.items())[:10]:
    print(genre, ":", count)

# Movies with the largest number of genres
top_multi_genre = movies.most_genres(10)
print("\nTop 10 movies by number of genres:")
for title, n_genres in top_multi_genre.items():
    print(f"{title} -> {n_genres} genres")

# time measurement for one of the methods
%timeit movies.dist_by_genres()
```

## Ratings (Movies perspective)

We look at how ratings are distributed over years and rating values, and which movies are most rated / highest rated / most controversial.

```python
# Create helper object for movie-based stats
movies_stats = Ratings.Movies(ratings)

# Distribution of ratings by year
ratings_by_year = movies_stats.dist_by_year()
print("Ratings count by year (first 10 years):")
for year, count in list(ratings_by_year.items())[:10]:
    print(year, ":", count)

# Distribution of rating values
ratings_by_value = movies_stats.dist_by_rating()
print("\nRatings distribution by value:")
for value, count in ratings_by_value.items():
    print(value, ":", count)

# Top movies by number of ratings
top_by_count = movies_stats.top_by_num_of_ratings(10)
print("\nTop 10 movies by number of ratings:")
for title, count in top_by_count.items():
    print(title, ":", count)

# Top movies by average rating
top_by_avg = movies_stats.top_by_ratings(10, metric="average")
print("\nTop 10 movies by average rating:")
for title, avg in top_by_avg.items():
    print(title, ":", avg)

# Top movies by median rating
top_by_median = movies_stats.top_by_ratings(10, metric="median")
print("\nTop 10 movies by median rating:")
for title, med in top_by_median.items():
    print(title, ":", med)

# Most controversial movies (highest variance)
top_controversial = movies_stats.top_controversial(10)
print("\nTop 10 most controversial movies (by rating variance):")
for title, var in top_controversial.items():
    print(title, ":", var)

# %timeit for one of the heavy methods
%timeit movies_stats.top_by_ratings(10, metric="average")
```

## Ratings (Users perspective)

Now we look at how actively users rate movies and how stable or variable their ratings are.

```python
users_stats = Ratings.Users(ratings)

# Distribution by number of ratings per user
dist_num_ratings = users_stats.dist_by_num_of_ratings()
print("Distribution of users by number of ratings:")
for num_ratings, num_users in list(dist_num_ratings.items())[:10]:
    print(num_ratings, "ratings ->", num_users, "users")

# Distribution of users by their average rating
dist_user_avg = users_stats.dist_by_user_metric(metric="average")
print("\nDistribution of users by average rating (sample):")
for avg, num_users in list(dist_user_avg.items())[:10]:
    print(avg, "->", num_users, "users")

# Top users by rating variance
top_users_var = users_stats.top_by_variance(10)
print("\nTop 10 users by rating variance:")
for user_id, var in top_users_var.items():
    print("User", user_id, "variance:", var)

%timeit users_stats.top_by_variance(10)
```

## Tags analysis

We analyze textual tags: their length, popularity, and specific words.

```python
# Tags with the most words
most_words_tags = tags.most_words(10)
print("Top 10 tags by number of words:")
for tag, word_count in most_words_tags.items():
    print(f"{tag} -> {word_count} words")

# Longest tags by characters
longest_tags = tags.longest(10)
print("\nTop 10 longest tags by characters:")
for tag in longest_tags:
    print(f"{tag} -> {len(tag)} characters")

# Intersection of most words and longest
intersection_tags = tags.most_words_and_longest(10)
print("\nTags that are both among the longest and with the most words:")
for tag in intersection_tags:
    print(tag)

# Most popular tags (by frequency)
popular_tags = tags.most_popular(10)
print("\nTop 10 most popular tags:")
for tag, count in popular_tags.items():
    print(f"{tag} -> {count} times")

# Example of tags containing a specific word
love_tags = tags.tags_with("love")
print("\nTags containing the word 'love':")
for tag in love_tags:
    print(tag)

%timeit tags.most_popular(10)
```

## External information via Links (IMDB-like)

Here we use the Links class to get external information such as directors, budgets, runtime, etc.

```python
# Example: get IMDB-like info for a few movies
sample_movie_ids = list(range(1, 6))
fields = ["title", "director", "budget", "gross", "runtime"]
imdb_rows = links.get_imdb(sample_movie_ids, fields)
print("IMDB info for some sample movies:")
for row in imdb_rows:
    print(row)

# Top directors by number of movies
top_dirs = links.top_directors(5)
print("\nTop 5 directors by number of movies:")
for director, count in top_dirs.items():
    print(director, ":", count)

# Most expensive movies
most_exp = links.most_expensive(5)
print("\nTop 5 most expensive movies:")
for title, budget in most_exp.items():
    print(title, ":", budget)

# Most profitable movies
most_prof = links.most_profitable(5)
print("\nTop 5 most profitable movies:")
for title, profit in most_prof.items():
    print(title, ":", profit)

# Longest movies by runtime
longest_movies = links.longest(5)
print("\nTop 5 longest movies:")
for title, runtime in longest_movies.items():
    print(title, ":", runtime, "minutes")

# Top cost per minute
top_cpm = links.top_cost_per_minute(5)
print("\nTop 5 movies by cost per minute:")
for title, cpm in top_cpm.items():
    print(title, ":", cpm)

%timeit links.top_directors(5)
```
