```python
import pandas as pd
```

## 1. Load JSON and set CarNumber as index

```python
# Read the JSON file from ex03/data
df = pd.read_json("data/auto.json", orient="records")

# Set CarNumber as the index
df = df.set_index("CarNumber")

df.head(), df.index.name, df.columns
```

## 2. Selections

```python
fines_gt_2100 = df[df["Fines"] > 2100]
fines_gt_2100.head()
```

```python
fines_gt_2100_refund_2 = df[
    (df["Fines"] > 2100) & (df["Refund"] == 2)
]
fines_gt_2100_refund_2.head()
```

```python
models_list = ["Focus", "Corolla"]

models_focus_corolla = df[df["Model"].isin(models_list)]
models_focus_corolla.head()
```

```python
car_list = [
    "Y7689C197RUS",
    "92928M178RUS",
    "7788KT197RUS",
    "H115YO163RUS",
    "X758HY197RUS",
]

selected_cars = df[df.index.isin(car_list)]
selected_cars
```

## 3. Aggregations with Make and Model

```python
median_fines_by_make = df.groupby("Make")["Fines"].median()
median_fines_by_make
```

```python
median_fines_by_make_model = df.groupby(["Make", "Model"])["Fines"].median()
median_fines_by_make_model
```

```python
count_fines_by_make_model = df.groupby(["Make", "Model"])["Fines"].count()
count_fines_by_make_model
```

```python
min_max_fines_by_make_model = df.groupby(["Make", "Model"])["Fines"].agg(["min", "max"])
min_max_fines_by_make_model
```

```python
std_fines_by_make_model = df.groupby(["Make", "Model"])["Fines"].std()
std_fines_by_make_model
```

## 4. Aggregations with car numbers

```python
fines_count_by_car = (
    df.groupby(level=0)["Fines"]
      .count()
      .sort_values(ascending=False)
)

fines_count_by_car.head(10)
```

```python
top1_by_count = fines_count_by_car.index[0]
top1_by_count, fines_count_by_car.iloc[0]

top1_by_count_rows = df.loc[top1_by_count]
top1_by_count_rows
```

```python
fines_sum_by_car = (
    df.groupby(level=0)["Fines"]
      .sum()
      .sort_values(ascending=False)
)

fines_sum_by_car.head(10)
```

```python
top1_by_sum = fines_sum_by_car.index[0]
top1_by_sum, fines_sum_by_car.iloc[0]

top1_by_sum_rows = df.loc[top1_by_sum]
top1_by_sum_rows
```

```python
car_model_counts = (
    df.reset_index()
      .groupby("CarNumber")["Model"]
      .nunique()
      .reset_index(name="n_models")
)

cars_with_multiple_models = car_model_counts[car_model_counts["n_models"] > 1]
cars_with_multiple_models
```

## See the list of models for each CarNumber

```python
car_models_table = (
    df.reset_index()
      .groupby("CarNumber")["Model"]
      .agg(["nunique", lambda x: sorted(set(x))])
      .reset_index()
)

car_models_table = car_models_table.rename(columns={"<lambda_0>": "models"})
car_models_table.head(10)
```
