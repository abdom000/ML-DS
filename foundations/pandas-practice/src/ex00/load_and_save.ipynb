pandas import

```python
import pandas as pd
```

Reading and cleaning the feed-views.log file

```python
df = pd.read_csv(
    "data/feed-views.log",
    sep="\t",
    header=None,
    names=["datetime", "user"],
    skiprows=[2, 3],
    skipfooter=2,
    engine="python"
)

df.head()
```

Convert datetime and use it as index

```python
df["datetime"] = pd.to_datetime(df["datetime"])

df = df.set_index("datetime")

df.index.name = "date_time"

df.head()
```

Save result to file ; separator

```python
df.to_csv(
    "data/feed-views-semicolon.log",
    sep=";",    
    index=True  
)
```
